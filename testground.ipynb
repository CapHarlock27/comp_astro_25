```python
import batman
import numpy as np
import matplotlib.pyplot as plt
```

```python
params = batman.TransitParams()
params.t0 = 0.                       #time of inferior conjunction
params.per = 32.939623                     #orbital period
params.rp = 0.0212                       #planet radius (in units of stellar radii)
params.a = 30.73                       #semi-major axis (in units of stellar radii)
params.inc = 89.5785                    #orbital inclination (in degrees)
params.ecc = 0.2                      #eccentricity
params.w = 354.3                       #longitude of periastron (in degrees)
params.u = [0.391617, 0.019183]                #limb darkening coefficients [u1, u2]
params.limb_dark = "quadratic"       #limb darkening model
```

```python
t = np.linspace(-0.05, 0.05, 100)
```

```python
m = batman.TransitModel(params, t)    #initializes model
flux = m.light_curve(params)          #calculates light curve
```

```python
plt.plot(t, flux)
plt.xlabel("Time from central transit")
plt.ylabel("Relative flux")
plt.show()
```
